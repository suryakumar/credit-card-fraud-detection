```python
import pandas as pd

df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Fraud distribution
print(df['Class'].value_counts())
```

```python
import matplotlib.pyplot as plt

#Amount distribution histogram by Fraud

df_good = df[df['Class']==0]
df_fraud = df[df['Class']==1]

#create histograms
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.hist(df_good['Amount'], bins=100, alpha=0.7)
plt.yscale('log')
plt.title('Valid Transactions')

plt.subplot(1,2,2)
plt.hist(df_fraud['Amount'], bins=100, alpha=0.7)
plt.yscale('log')
plt.title('Fraud Transactions')
plt.tight_layout()
plt.show()
```

```python
#Correlation Heatmaps

import seaborn as sns

overall_corr = df.corr()
good_corr = df_good.corr()
fraud_corr = df_fraud.corr()

# Creating heatmaps
plt.figure(figsize=(12, 18))

# Overall Correlation Heatmap
plt.subplot(3, 1, 1)
sns.heatmap(overall_corr, annot=False, cmap='coolwarm')
plt.title('Overall Correlation')

# Correlation Heatmap for Binary Variable = 0
plt.subplot(3, 1, 2)
sns.heatmap(good_corr, annot=False, cmap='coolwarm')
plt.title('Valid Transactions Correlation Matrix')

# Correlation Heatmap for Binary Variable = 1
plt.subplot(3, 1, 3)
sns.heatmap(fraud_corr, annot=False, cmap='coolwarm')
plt.title('Fraud Transactions Correlation Matrix')

plt.tight_layout()
plt.show()
```
